```python
import pandas as pd
```

```python
df = pd.read_csv('cmucorr.csv')
df.head()
```

```python
dfr = df[[column for column in list(df.columns) if not column.startswith('Unnamed')]]
dfr.head()
```

```python
names = list(dfr.columns)
wnames = [name for name in names if name.endswith('weighted')]
wnames
```

```python
var2id = {var:id for (id, var) in enumerate(names) if var in wnames}
var2id # all indices from 49 to 97h
```

```python
reqd = ['pct_cmnty_cli_weighted', 'pct_self_anosmia_ageusia_weighted', 'pct_hh_cli_weighted',
        'pct_hh_fever_weighted', 'pct_self_fever_weighted', 'pct_hh_sore_throat_weighted',
        'pct_avoid_contact_all_or_most_time_weighted', 'pct_hh_difficulty_breathing_weighted',
        'pct_self_persistent_pain_pressure_in_chest_weighted', 'pct_self_runny_nose_weighted',
        'pct_worked_outside_home_weighted',  'pct_tested_and_positive']
reqd2id = {var: i for (i, var) in enumerate(names) if var in reqd}
reqidx = [i for (i, var) in enumerate(names) if var in reqd]
reqd2id, reqidx
```

```python
graph = {
    "nodes": [
        
    ],
    "edges": [
        
    ]
}
graph["nodes"] = [{"id":i, "name": var} for (i, var) in enumerate(reqd)]
graph["nodes"]
```

```python
matrix = list(dfr.to_records(index=False))
```

```python
graph["edges"] = []
import itertools
for i,j in itertools.product(reqidx, reqidx):
    weight = matrix[i][j]
    source_n, target_n = names[i], names[j]
    source = 0
    target = 0
    for node in graph["nodes"]:
        if node["name"] == source_n:
            source = node["id"]
        if node["name"] == target_n:
            target = node["id"]
            
    # print(source_n, source, i, target_n, target, j, weight)
    if source == target:
        continue
    graph["edges"].append({
        "source":source,
        "target":target,
        "weight": weight
    })

graph["edges"]
        
```
