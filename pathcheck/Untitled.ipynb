```python
import pandas as pd
```

```python
df = pd.read_csv('overall-state.csv')
list(df.columns.values)
```

```python
df2  = df[['state_code', 'gender', 'age_bucket', 'pct_tested_and_positive_weighted']]
df2.head()
```

```python
data_overall18 = df2.loc[(df2['gender'] == 'overall') & (df2['age_bucket'] == '18-34')][['state_code', 'pct_tested_and_positive_weighted']]
data_overall18 = data_overall18.groupby(data_overall18['state_code']).aggregate({'pct_tested_and_positive_weighted':'mean'})
data_overall18.to_csv('overall-18.csv')
data_overall18
```
